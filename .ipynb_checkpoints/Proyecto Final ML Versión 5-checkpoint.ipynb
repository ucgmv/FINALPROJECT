#  Modelado de datos

## Normalización de los datos

```python
DataSet_norm = DataSet
to_normalize = ['age', 'trtbps', 'chol', 'thalachh', 'oldpeak'] #aca puse algunas, hay que ver cuales queremos normalizadas
for x in to_normalize:
    DataSet_norm[x] = (DataSet_norm[x]-DataSet_norm[x].mean())/DataSet_norm[x].std()
```

```python
# Separo en un vector y el output y en una matriz los X
y = DataSet_norm['output']
print(y.head())
```

```python
X = DataSet_norm.drop('output', axis=1)
print(X.head())
```

```python
# Separo el 20% de los datos para test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print('Tamaño de train_dataset:', X_train.shape, y_train.shape)
print('Tamaño de test_dataset:', X_test.shape, y_test.shape)
```

## Funciones para el modelo

```python
## Vamos a generar una matriz de confusión para poder visualizar los tipos de errores generados
def confu(y_test, y_pred):
    plt.figure(figsize=(7,3))
    cm = confusion_matrix(y_test, y_pred)
    class_label = ['Alto Riesgo', 'Bajo Riesgo']
    DataSet_cm = pd.DataFrame(cm, index=class_label, columns =class_label)
    sns.heatmap(DataSet_cm, annot =True, cmap = 'OrRd', linewidths = 2, fmt='d')
    plt.title("Matriz de Confusión", fontsize = 15)
    plt.xlabel('Predecido')
    plt.ylabel('Real')
    plt.show()
            
    acc=accuracy_score(y_test,y_pred)
    mse=mean_squared_error(y_test, y_pred)
    precision, recall, fscore, train_support = score(y_test, y_pred, pos_label=1, average='binary')
    print('Precision: {:.4f}% \nRecall: {:.4f}% \nF1-Score: {:.4f}% \nAccuracy: {:.4f}% \nMean Square Error: {:.4f}'.format(precision*100, recall*100, fscore*100, acc*100, mse))
    
    
   # CURVA ROC número 1, coherente con lo visto en clase
    sk.metrics.roc_curve(y_test, y_pred, pos_label=None, sample_weight=None, drop_intermediate=True)
    fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred)
    roc_auc = metrics.auc(fpr, tpr)
    display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc,estimator_name='example estimator')
    display.plot()  
    plt.title('Curva Roc')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.show() 
    
    
    #CURVA ROC número 2, no condice con lo visto en clase
    from sklearn.metrics import roc_curve
    # calculating the probabilities
    y_pred_prob = modelo_lr.predict_proba(X_test)[:,1]

    # instantiating the roc_cruve
    fpr,tpr,threshols=roc_curve(y_test,y_pred_prob)

    # plotting the curve
    plt.plot([0,1],[0,1])
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.plot(fpr,tpr,label='Logistic Regression')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.show()
```

```python
def create_baseline():
    model = Sequential()
    model.add(Dense(13, input_dim=13, activation='sigmoid'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

## Cross-Validation

```python
modelos = {}
```

```python
def cross_validation(modelo, nombre_modelo):
    cv = KFold(n_splits=10, random_state=1, shuffle=True)
    # evaluate model
    scores = cross_val_score(modelo, X_train, y_train, scoring='recall', cv=cv, n_jobs=-1)
    # los agrego para compararlos
    modelos[nombre_modelo] = [mean(scores).item(), std(scores).item(), len(scores)]
```

```python
# Este test set no se va a tocar hasta despues del cross validation. Una vez sacado el mejor, vamos a probar contra este.
y = DataSet_norm['output']
X = DataSet_norm.drop('output', axis=1)
print(X.head())
print(y.head())
```

```python
#Regresión Logística
modelo = LogisticRegression()
cross_validation(modelo, "Regresión Logística")
```

```python
#KNN
modelo = KNeighborsClassifier(n_neighbors=21)
cross_validation(modelo, "KNN")
```

```python
#SVM
modelo = SVC()
cross_validation(modelo, "SVM")
```

```python
#Arbol de Decisión
modelo = DecisionTreeClassifier()
cross_validation(modelo, "Arbol de decisión")
```

```python
x = []
y = []
errores = []
print('\033[1m')
print(" "*15+"Recall", end="\n\n")
print("{:<20} {:<10} {}".format('Modelo','Media','Desvío'),end="")
print('\033[0m')
for nombre, estadisticos in modelos.items():
    x.append(nombre)
    media = estadisticos[0]
    desvio = estadisticos[1]
    n = estadisticos[2]
    y.append(media)
    z = stats.norm.ppf(0.975)
    error = z * desvio / sqrt(n)
    errores.append(error)
    print("{:<20} {:.2f}{:<5} {:.2f}{}".format(nombre, media*100, "%", desvio*100, "%"))
    
print()
plt.figure()
plt.errorbar(x, y, yerr=errores, fmt = 'o', color = 'k', capsize=5)
plt.title('Comparación de modelos')
plt.ylim([0.6, 1.00])
plt.show()
```
